```
from google.colab import drive
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

drive.mount('/content/drive')

def convolution2D(image2D, kernel3x3):
  convolved2D = np.zeros_like(image2D)
  imaged_paded = np.zeros((image2D.shape[0] + 2, image2D.shape[1] + 2))
  imaged_paded[1:-1, 1:-1] = image2D
  for i in range(image2D.shape[1]):
    for j in range(image2D.shape[0]):
      convolved2D[j, i]= (kernel3x3*imaged_paded[j:j+3, i:i+3]).sum()
  return convolved2D

image2D = np.loadtxt('/content/drive/My Drive/Colab Notebooks/once.txt', delimiter=',')
sns.heatmap(image2D, cmap='Greens')
plt.show()

edge_detect_filter_3x3 = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])

# Convolve once
convolved_image = convolution2D(image2D, edge_detect_filter_3x3)

sns.heatmap(convolved_image, cmap='Greens')
plt.show()

# Convolve again
convolved_image = convolution2D(convolved_image, edge_detect_filter_3x3)

sns.heatmap(convolved_image, cmap='Greens')
plt.show()

print(image2D.shape[0])
print(image2D.shape[1])
```

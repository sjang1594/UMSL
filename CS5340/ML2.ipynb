```
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5, 10, 11, 12, 13])
y = np.array([0, 0, 0, 1, 1, 1, 1, 1, 1])

plt.plot(x, y, 'r*')
```

```
X_mean = np.mean(x)
Y_mean = np.mean(y)
num = 0
den = 0
for i in range(len(x)):
  num +=(x[i] - X_mean)*(y[i] - Y_mean)
  den +=(x[i] - X_mean)**2

#To get the regression coefficient -> wo + w1*x = y^
least_square = num/den                #This will be the w1
c = Y_mean - least_square*X_mean      #This will be the w0

print(least_square, c)
```

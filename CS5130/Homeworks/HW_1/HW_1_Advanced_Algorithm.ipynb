## **Approach -2 : Plot**

### **Entire for ranking function**

```
#plot
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import gamma
```

```
x = np.linspace(1, 50, 100)
plt.plot(x, np.sqrt(x), label="sqrt(x)")
plt.plot(x, np.square(x), label = "x^2")
plt.plot(x, gamma(x), label="x!")
plt.plot(x, gamma(np.log2(x)), label="(lgx)!")
plt.plot(x, np.log2(x)**2, label="(lgx)^2")
plt.plot(x, np.log2(gamma(x)), label="lg(x!)")
plt.plot(x, np.power(2, (np.power(2,x))), label="2^2^x")
plt.plot(x, np.ones(100)*2, label="2")
plt.plot(x, x * np.power(2,x), label="x*2^x")
plt.plot(x, x ** np.log2(np.log2(x)), label="x^lg(lg(x))")
plt.plot(x, np.log(x), label="ln(x)")
plt.plot(x, np.ones(100), label="1")
plt.plot(x, np.exp(x), label="e^x")
plt.plot(x, gamma(x+1), label="(x+1)!")
plt.plot(x, np.sqrt(np.log2(x)), label="sqrt(lgx)")
plt.plot(x, x, label="x")
plt.plot(x, np.power(2,x), label="2^x")
plt.plot(x, x * np.log2(x), label="x*lg(x)")
plt.plot(x, np.power(2, np.power(2,(x+1))), label = "2^2^x+1")

plt.title("Time Complexity VS Input Data")
plt.xlabel("input data(x)")
plt.ylabel("time complexity")
plt.axis([0, 20, 0, 500])
plt.legend(loc="upper center", bbox_to_anchor=(1.2, 1.2), shadow=True, ncol=1)
plt.show()
```

```
x = np.linspace(1, 50, 100)
plt.plot(x, np.power(2, np.power(2,(x+1))), label = "2^2^x+1")
plt.plot(x, np.power(2, (np.power(2,x))), label="2^2^x")
plt.plot(x, x * np.power(2,x), label="x*2^x")
plt.plot(x, gamma(x+1), label="(x+1)!")
plt.plot(x, gamma(x), label="x!")
plt.plot(x, np.exp(x), label="e^x")
plt.plot(x, np.power(x, np.log2(np.log2(x))), label="x^lg(lg(x))")

plt.title("Time Complexity VS Input Data - Subplot")
plt.xlabel("input data(x)")
plt.ylabel("time complexity")
plt.axis([0, 17, 0, 5000])
plt.legend(loc="upper center", bbox_to_anchor=(1.2, 1.2), shadow=True, ncol=1)
plt.show()
```

### **Result on High Rank**
* By using the stirling approximation on (lgn)!, we can approximate the values for (lgn)!

I used this resource to find out the rank of lgn!

Here is the link from Stack overflow
* [log(n)!](https://stackoverflow.com/questions/47068377/o-log-n-and-o-log-n)
* [log(n)! -2](https://stackoverflow.com/questions/40067183/is-olog-n-polynomial)

The summary is telling that (lgn)! function will be higher rank than any polynomial function, such as ${x^2}$ Also, it mentions that lg(n!) should be lower than any polynomial function.

So then, **Rank will be the following, $2^{2^{x+1}}$, $2^{2^{x}}$, (x+1)!, (x)!, $x*2^x$, $x^{lg(lg(x))}$, (lgx)!, ${x^2}$
